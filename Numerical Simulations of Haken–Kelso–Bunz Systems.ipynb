## Mathematical Introduction

The Haken–Kelso–Bunz (HKB) model approximates the evolution of a system of non-linearly coupled, non-linear oscillators.  There are a couple levels at which the approximation can apply.  The more general level evolves both the amplitudes and the phases of the oscillators' motions.  However, the phases have recieved more attention historically, and one can specialize the equations further to evolve only those variables, with the amplitudes held fixed.

In the more general framework, the instantaneous state of each oscillator can be encoded in a single, complex variable $z_i(t) := r_i(t)\, \mathrm{e}^{\mathrm{i} \theta_i(t)}$, which is connected to the trajectory $x_i(t)$ of that oscillator by 

\begin{equation}
    x_i(t) = \operatorname{Re} \bigl[ z_i(t)\, \mathrm{e}^{\mathrm{i} \Omega t} \bigr] 
    \qquad\text{and}\qquad 
    \dot x_i(t) = - \Omega \operatorname{Im} \bigl[ z_i(t)\, \mathrm{e}^{\mathrm{i} \Omega t} \bigr].
\end{equation}

The parameter $\Omega$ is arbitrary at this point, and will be chosen later to simplify the collective dynamics of the oscillators.  Note, in particular, that any oscillator moving harmonically with frequency $\Omega$ will have $z_i(t) = \text{const.}$

The evolution of the complex variables $z_i(t)$ in the HKB model derives from the equations of motion 

\begin{equation}
    \ddot x_i - \bigl( \alpha_i - \beta_i\, \dot x_i^2 - \gamma_i\, \dot x_i^2 \bigr)\, \dot x_i + \omega_i^2\, x_i  
        = - \sum_{j \ne i} \bigl( a_{ij} - b_{ij}\, (\dot x_i - \dot x_j)^2 - c_{ij}\, (x_i - x_j)^2 \bigr) \bigl( \dot x_i - \dot x_j \bigr). 
\end{equation}

The left side of this equation describes the *autonomous* dynamics of a (hybrid, Rayleigh–van der Pol) non-linear oscillator with coordinate $x_i(t)$.  Note that, in the absence of the coupling terms on the right, the constant term in the damping factor has the "wrong" sign if $\alpha_i > 0$.  That, is, it describes an inherent, continual *excitation* of the oscillator.  The other two terms in the damping factor have the opposite sign, however, and will properly damp the oscillator's motion once its amplitude is large enough.  The net result is that the autonomous dynamics of a single oscillator will converge on a non-trivial, stable motion (unlike the damped linear oscillator, for example, which evolves to a state of rest).

The coupling terms in the previous equation serve to perturb the stable, autonomous evolution of each oscillator, depending on the configurations of the other oscillators in the ensemble.  Note that, in contrast to the autonomous terms, the constant term in the coupling factor has a damping effect if $a_{ij} > 0$, while the non-linear terms are excitatory if $b_{ij} > 0$ and $c_{ij} > 0$.

Substituting for $x_i(t)$ and its derivatives in the oscillator equations of motion in terms of the complex variables $z_i(t)$ defined above yields a complicated evolution equation with explicit factors of $\mathrm{e}^{\mathrm{i} \Omega t}$ in its coefficients.  However, the goal of the HKB model is to describe collective states of motion of the oscillators wherein each moves nearly harmonically with frequency $\Omega$.  For such motions, it makes sense to approximate the evolution equations by their average over a single period of that presumed, harmonic oscillation.  Doing so leads to the approximate evolution equations 

\begin{equation}
    \dot z_i 
        = \biggl( \frac{\alpha_i}{2} - \frac{3 \beta_i \Omega^2 + \gamma_i}{8}\, |z_i|^2 + \mathrm{i}\, \frac{\omega_i^2 - \Omega^2}{2 \Omega} \biggr) z_i 
            - \sum_{j \ne i} \biggl( \frac{a_{ij}}{2} - \frac{3 b_{ij} \Omega^2 + c_{ij}}{8}\, |z_i - z_j|^2 \biggr) \bigl( z_i - z_j \bigr).
\end{equation}

This is the more general form of the HKB equations that we will consider below.

Writing $z_i(t) = r_i(t)\, \mathrm{e}^{\mathrm{i} \theta_i(t)}$, we can separate the the complex form of the HKB equations given above into separate evolutions equations for the amplitudes and phases of the oscillators.  The phase equations are 

\begin{equation} 
    \dot\theta_i 
        = \operatorname{Im} \frac{\dot z_i}{z_i} 
        = \frac{\omega_i^2 - \Omega^2}{2 \Omega} 
            - \sum_{j \ne i} A_{ij} \sin (\theta_i - \theta_j) 
            - \sum_{j \ne i} B_{ij} \sin 2 (\theta_i - \theta_j)
    \qquad\text{with}\qquad 
    A_{ij} := \frac{a_{ij}}{2}\, \frac{r_j}{r_i} - B_{ij}\, \frac{r_i^2 + r_j^2}{r_i r_j} 
    \qquad\text{and}\qquad 
    B_{ij} := \frac{3 b_{ij} \Omega^2 + c_{ij}}{8}\, r_j^2.
\end{equation}

Meanwhile, the amplitude equations are 

\begin{equation} 
    \frac{\dot r_i}{r_i} 
        = \operatorname{Re} \frac{\dot z_i}{z_i} 
        = \frac{\tilde\alpha_i}{2} - \frac{3 \tilde\beta_i + \tilde\gamma_i}{8}\, r_i^2 
            + \sum_{j \ne i} 2 B_{ij} 
            + \sum_{j \ne i} \frac{A_{ij}\, r_j - 2 B_{ij}\, r_i}{r_j} \cos (\theta_i - \theta_j) 
            + \sum_{j \ne i} B_{ij} \cos 2 (\theta_i - \theta_j) 
    \qquad\text{with}\qquad 
    \tilde\alpha_i := \alpha_i - \sum_{j \ne i} a_{ij}, 
\end{equation}

and $\tilde\beta_i$ and $\tilde\gamma_i$ defined similarly.

A further approximation to the HKB equations supposes that the amplitudes of the oscillators' motions evolve much more slowly than their (anomalous) phases.  In this further approximation, we may treat $A_{ij}$ and $B_{ij}$ as constants in the phase evolution equations given above, and drop the amplitude evolution equations altogether.  This reduced system of phase evolution equations is what is most commonly called the "HKB model."

## Code Initialization

We will need the following libraries and routines preloaded.

```python
import numpy as np
import matplotlib.pyplot as plt

# The SciMath project includes packages to support scientific and mathematical calculations. 
from numpy.lib import scimath as npc

# This function numerically integrates a system of ordinary differential equations given an initial value.
from scipy.integrate import solve_ivp
```

We will need an instance of the built-in random number generator to create initial data sets and diverse parameter values for systems of non-identical oscillators.

```python
myRNG = np.random.default_rng(seed = 100)
```

The following routine computes the time derivative of the HKB phases from given values of the phases themselves.  Its form matches that assumed by the built-in integrator for initial-value problems.  It presumes that the frequencies $\omega_i$ and $\Omega$ have been pre-defined, as well as the reduced coupling constants $A_{ij}$ and $B_{ij}$.

```python
def HKBphase(t, theta) : 
    
    dot_theta = np.empty_like(theta)
    
    for i in range(len(theta)) : 
        dot_theta[i] = omega[i] - A[i,:] @ np.sin(theta[i] - theta[:]) - B[i,:] @ np.sin(2 * (theta[i] - theta[:]))
    
    return dot_theta
```

The following defines the lower limit of the visible window in the vertical axis in the plots generated below. 

```python
ymin  = -0.8
```

## System of Identical Oscillators

First we define the number $N$ of oscillators, the time $T$ at which to end our evolution, and the number of time points we wish to simulate $N_T$.

```python
N = 3
T = 10
N_T = 4000
```

```python
omega = np.array((1, 10, 20))
Omega = np.sqrt(np.mean(omega**2))
```

```python
A = np.array([[0, 1, 5], [1, 0, 5], [5, 5, 0]])
B = np.array([[0, 1, 1], [1, 0, 1], [1, 1, 0]])
```

```python
theta0 = myRNG.uniform(-np.pi, np.pi, size = N)
sol = solve_ivp(HKBphase, (0, T), theta0, dense_output = True, max_step = 0.05)
```

```python
# Create the lattice of time points by dividing the interval (0, T) into N_T time points.
t = np.linspace(0, T, N_T)

# Evaluate the numerical solution at this lattice of time points.
theta  = sol.sol(t)

# Take the transpose of the solution so that the time points are on the x axis and the phi points on the y axis.
#phi = np.transpose(phi - phi[0])
theta = np.transpose(theta)

# Produce a plot.
fig, ax = plt.subplots(figsize = (12, 6))

# Set a colormap for the different oscillators.  Choose colormap in jet mode. 
colormap = plt.cm.jet

# Set the colors for the map by dividing the spectrum into N pieces, one for each oscillator.
colors = [colormap(c) for c in np.linspace(0., 1., N)]

# Set the colors to use?
ax.set_prop_cycle('color', colors)


for s in range(np.ceil((ymin - np.amax(theta) / np.pi) / 2).astype(int), np.floor((ymin - np.amin(theta) / np.pi) / 2).astype(int) + 3) : 
    ax.plot (t, theta / np.pi + 2 * s) 

ax.set(xlabel = 't', ylabel = 'theta / pi', ylim = [ymin, ymin + 4.])
ax.legend(np.char.add('theta', np.arange(0, N).astype(str)), bbox_to_anchor = (1.05, 1), loc = 'upper left', borderaxespad = 0.);
```

## Systems of Three Oscillators with Nonuniform Coupling and Frequency Differences

### Potential Function Methods for N-oscillator HKB System

For a system of N oscillators, the equation for the absolute phase dynamics takes the form 

\begin{equation}
\dot \theta_i = \omega_i - \sum_{i \neq j}(A_{ij}\sin(\theta_i - \theta_j) + B_{ij}\sin2(\theta_i - \theta_j))
\end{equation}

We want to recast this in terms of a potential function $V(\boldsymbol{\theta})$ such that 

\begin{equation}
\boldsymbol{\dot \theta} = -\nabla V, \textrm{i. e.} \left ( \dot \theta_1, ..., \dot \theta_N \right ) = \left (-\frac{\partial V}{\partial \theta_1}, ..., -\frac{\partial V}{\partial \theta_N} \right )
\end{equation}

Such a function exists if and only if 

\begin{equation}
\nabla \times (-\nabla V) = \nabla \times \boldsymbol{\dot\theta} = \mathbf{0}
\end{equation}

as must have be the case for any pure gradient function.  Plugging in the equation for $\boldsymbol{\dot \theta}$, we find this condition is met if and only if

\begin{equation}
A_{ij} = A_{ji}, B_{ij} = B_{ji}
\end{equation}

i. e. we can derive this dynamical system from a scalar potential function only for systems of oscillators with symmetric coupling terms.


Now, the potential functioning satisfying these conditions is

\begin{equation}
V(\boldsymbol{\theta}) = -\boldsymbol{\omega} \cdot \boldsymbol{\theta} - \frac{1}{2} \sum_{i, j, i \neq j}(A_{ij}\cos(\theta_i - \theta_j) + \frac{1}{2}B_{ij}\cos2(\theta_i - \theta_j))
\end{equation} where $\boldsymbol{\omega} = (\omega_1, ..., \omega_N)$.

For a system of three oscillators, this becomes

\begin{equation}
V(\theta_1, \theta_2, \theta_3) = -\omega_1 \theta_1 - \omega_2 \theta_2 - \omega_3 \theta_3 - A_{12}\cos(\theta_1 - \theta_2) - A_{13}\cos(\theta_1 - \theta_3) - A_{23}\cos(\theta_2 - \theta_3) - \frac{1}{2}(B_{12}\cos2(\theta_1 - \theta_2) + B_{13}\cos2(\theta_1 - \theta_3) + B_{23}\cos2(\theta_2 - \theta_3))
\end{equation}

This is a 4-dimensional system, i.e. a function of three variables.  Such a system can be hard to visualize in its present form.  However, we can recast this into a 3-dimensional system by changing variables from the absolute phases to the relative phases.  Let

\begin{equation}
\phi := \theta_1 - \theta_2, \psi = \theta_1 - \theta_3, \psi - \phi = \theta_2 - \theta_3
\end{equation}

Also, define

\begin{equation}
\delta\omega_\phi := \omega_1 - \omega_2, \delta\omega_\psi := \omega_1 - \omega_3, \delta\omega_d := \omega_2 - \omega_3
\end{equation}

and 

\begin{equation}
A_{\phi} := A_{12}, A_{\psi} := A_{13}, A_d := A_{23}, B_{\phi} := B_{12}, B_{\psi} := B_{13}, B_d := B_{23}
\end{equation}

Then, in terms of the relative phases, the potential function becomes

\begin{equation}
V(\phi, \psi) = -\delta \omega_\phi \phi -\delta \omega_\psi \psi - A_\phi \cos\phi - A_\psi \cos\psi - A_d \cos(\phi - \psi) - B_\phi \cos2\phi - B_\psi \cos2\psi - B_d \cos2(\phi - \psi)
\end{equation}

This gives us a 3-dimensional system we can easily visualize to identify the key stability features of the dynamical system.  Valleys in the potential landscape correspond to stable fixed points of the system.

The valleys in the potential landscape of the above potential function follow a distinct pattern as we vary the coupling parameters.  First, if all pairs of coupling constants are subcritical (i. e. for all pairs of coupling terms $(A_{ij}, B_{ij})$ with values $\frac{B_{ij}}{A_{ij}} > \frac{1}{2}$), there is a local minimum about $\phi = \psi = 0$ which corresponds to the state where all oscillators are in-phase, local minima about $\phi = \pm \pi$ and $\psi = 0$ which corresponds to the state where oscillators 1 and 3 are in-phase and oscillator 2 antiphase to the other two, local minima about $\psi = \pm \pi$ and $\phi = 0$ which corresponds to the state where oscillators 1 and 2 are in-phase and oscillator 3 antiphase to the other two, and local minima about $\phi = \pm \pi$ and $\psi = \pm \pi$ which corresponds to the state where oscillators 2 and 3 are in-phase with each other and both antiphase to oscillator 1.

Then, if we increase one of the $A_{ij}$ such that one pair of coupling constants $(A_{ij}, B_{ij})$ becomes supercritical, one of two things will occur.  Either the diagonal antiphase minima (i. e. the state where oscillator 1 is antiphase to both oscillators 2 and 3) will be lost if the other pairs of coupling ratios are small enough, or, if the other pairs of coupling ratios are large enough, the diagonal antiphase state will still exist but it will be less deep of a valley, and hence the stability will be reduced.

Next, if we make two pairs of coupling constants supercritical, a similar pattern will follow: the diagonal antiphase states will be lost completely, and either the vertical or horizontal antiphase states will remain with reduced stability if the ratio of the diagonal coupling constants is far enough from criticality, or, if this ratio is close to criticality, the horizontal and vertical antiphase states will be lost.  

Finally, if all three pairs of coupling constant ratios become supercritical, all of the local minima corresponding to the antiphase states will be lost and the only remaining stable fixed point will be the state where all oscillators are in-phase with each other.

These are the predictions made by examining the above potential function.  We will now simulate the N = 3 HKB oscillator system and examine the fixed point states which result for particular pairs of coupling parameters, and see that the resulting behavior does indeed follow the pattern predicted by the potential function method.

### Eigenvalues and Stability

By computing the Hessian 

\begin{equation} 
H_{\mu \nu} = \frac{\partial^2 V}{\partial \phi^\mu \partial \phi^\nu} 
\end{equation}

of the potential function $V(\phi^\mu)$, and multiplying it by the metric tensor 

\begin{equation}
g^{\mu \nu} = 
\begin{pmatrix}
2 & 1 \\ 1 & 2
\end{pmatrix}
\end{equation}

for transforming from the $\theta^\mu$ coordinates to the $\phi^\mu$ coordinates, we obtain the system Jacobian 

\begin{equation}
J_\nu^\mu = g^{\mu \rho}H_{\rho \nu}
\end{equation}  

Then, by computing the eigenvalues of the Jacobian evaluated at a fixed point $\phi^\mu_{eq}$, we can determine if the fixed point is stable.  If all the eigenvalues of the Jacobian are positive, then the fixed point is stable.  Conversely, if any of the eigenvalues are negative, then the fixed point is unstable.  This places constraints on the possible values the coupling parameters $A_i$ and $B_i$ can take for restoring stability to fixed points that would otherwise be unstable without the coupling to the third oscillator.

#### Diagonal Antiphase

For the fixed point $\phi^\mu_{eq} = (\phi_1 = \pi, \phi_2 = \pi)$, the eigenvalues are given by \begin{equation}
\lambda_{\pm} = (a + b + c) \pm \sqrt{(a + b + c)^2 - 3(ab + (a + b)c)}
\end{equation}

where $a := -A_1 + 2B_1, b := -A_2 + 2B_2,$, and $c := A_d + 2B_d$.  

Therefore, in order for $\phi^\mu_{eq} = (\pi, \pi)$ to be stable, two conditions must be met.  \begin{equation}i. -(A_1 + A_2 - A_d) + 2(B_1 + B_2 + B_d) > 0 
\end{equation} and \begin{equation}ii. (-A_1 + 2B_1)(-A_2 + 2B_2) + (-(A_1 + A_2) + 2(B_1 + B_2))(A_d + 2B_d) > 0
\end{equation}  

If both of these conditions are satisfied, then the fixed point is stable.

We will write a function to compute these two quantities given values of $A_i$, $B_i$ to determine if these conditions are satisfied, and will see how these calculations compare with the potential function to test for consistency.

```python
def eval_diag(A_1, A_2, A_d, B_1, B_2, B_d):
    
    condition_1 = -(A_1 + A_2 - A_d) + 2*(B_1 + B_2 + B_d)
    
    print(condition_1)

    condition_2 = (-A_1 + 2*B_1)*(-A_2 + 2*B_2) + (-(A_1 + A_2) + 2*(B_1 + B_2))*(A_d + 2*B_d)
    
    print(condition_2)

    if (condition_1 > 0) & (condition_2 > 0):
        print('Fixed point is stable.')
    else:
        print('Fixed point is unstable')
```

## Extended HKB Model with Frequency Differences

### Simple HKB Model with Maximal Symmetry

The original HKB model was formulated with two nonlinearly coupled nonlinear oscillators with identical intrinsic eigenfrequencies, where the relative phase of the two oscillators evolves according to the equation

\begin{equation}
    \dot \phi = -a\sin\phi - 2b\sin2\phi. 
\end{equation}  

where $\phi$ is the relative phase.  The resulting dynamics evolve with in-phase and antiphase fixed points.  The in-phase fixed point always exists, while the in-phase and anti-phase fixed points coexist for  

\begin{equation}
    \frac{b} {a} > \frac {1} {4}.
\end{equation} 

This presence of both in-phase and antiphase fixed points is called multistability.

We will simulate the dynamics of the HKB model for different parameter values to familiarize with the model and see that it works how we would expect.

The following function gives the equation of motion for the HKB model for a given set of parameter values.

```python
def HKB(t, phi):
    
    dot_phi = -a*np.sin(phi) - 2*b*np.sin(2*phi)
    
    return dot_phi
```

Then we set the parameters $a$ and $b$, and the duration of the experiment T.

```python
a = 0.105
b = 0.105
T = 50
```

Now set the initial phases and integrate the system using RK4.

```python
phi0 = myRNG.uniform(0, 2*np.pi, size = 1) 
sol1 = solve_ivp(HKB, (0, T), phi0, dense_output = True)
```

Finally, plot the system for a set of time points.

```python
# Create the lattice of time points by dividing the interval (0, T) into N_T = 1000 time points.
t = np.linspace(0, T, 1000)

# Evaluate the numerical solution at this lattice of time points.
phi1 = sol1.sol(t)
phi1 = np.transpose(phi1)

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(t, phi1)

#ax.set(xlabel = 't', ylabel = 'phi / pi', ylim = [0, 4*np.pi])
```

However, if we decrease the ratio of $a$ and $b$ such that $ \frac{b} {a} < \frac {1} {4}$, say by taking $a = 5$ and $b = 1$, the antiphase fixed point should be lost.

### Breaking the Symmetry in Frequency

Now, consider the extended HKB model.  The equation for the phase dynamics of the extended HKB model modified as

\begin{equation}
    \dot \phi = \delta\omega -a\sin\phi - 2b\sin2\phi. 
\end{equation}

This is the same as the ordinary HKB model, except now we allow for a difference in the eigenfrequencies of the two oscillators, given by 

\begin{equation}
    \delta\omega = \omega_1 - \omega_2. 
\end{equation}  

This breaks the symmetry of the system, and alters the phase dynamics.  If we fix the coupling parameters $a$ and $b$, then for small values of $\delta\omega$ both the in-phase and antiphase fixed points still exist.  However, if we increase $\delta\omega$, the in-phase and antiphase fixed points cease to be stable, first giving way to metastability, and then as $\delta\omega$ is increased even further stability is lost completely.

Define a function for the extended HKB model.

```python
def EHKB(t, phi):
    
    dot_phi = d_omega -a*np.sin(phi) - 2*b*np.sin(2*phi)
    
    return dot_phi
```

Now set the values for the model parameters.

```python
a = 1
b = 1
d_omega = 1
```

Then, integrate the system on the same initial dataset.

```python
sol3 = solve_ivp(EHKB, (0, T), phi0, dense_output = True)
```

Finally, plot the results.

```python
# Create the lattice of time points by dividing the interval (0, T) into N_T = 1000 time points.
t = np.linspace(0, T, 1000)

# Evaluate the numerical solution at this lattice of time points.
phi1 = sol3.sol(t)
phi1 = np.transpose(phi1)

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(t, phi1)

#ax.set(xlabel = 't', ylabel = 'phi / pi', ylim = [0, 4*np.pi])
```

We see that a frequency difference $\delta\omega = 1$ is not large enough to alter the dynamics significantly.  Try cranking up the frequency difference to, say, $\delta\omega = 10$.

```python
d_omega = 10

sol4 = solve_ivp(EHKB, (0, T), phi0, dense_output = True)

# Create the lattice of time points by dividing the interval (0, T) into N_T = 1000 time points.
t = np.linspace(0, T, 1000)

# Evaluate the numerical solution at this lattice of time points.
phi1 = sol4.sol(t)
phi1 = np.transpose(phi1)

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(t, phi1)

#ax.set(xlabel = 't', ylabel = 'phi / pi', ylim = [0, 4*np.pi])
```

Now, both the in-phase and antiphase fixed points are lost and the relative phase simply grows linearly.  There should be a range of values for $\delta\omega$, not too large and not too small, such that the system exhibits metastability.  Let's try to identify a value of $\delta\omega$ within this range.

### Identifying Fixed Points and Stability with Potential Functions

A useful way to represent dynamical systems is with potential functions.  Generally speaking, in dynamical systems a potential function is some function $V(x)$ such that 

\begin{equation}
    \dot x = -\frac{dV}{dx}
\end{equation}

where $x$ is the state variable we wish to solve for.

For the extended HKB model, the potential of the system is 

\begin{equation}
    V(\phi) = -\delta\omega\phi -a\cos\phi - b\cos2\phi
\end{equation}

The benefit of the potential formulation is that one can readily identify fixed points and their stability by identifying points in the potential where $\frac{dV}{dx} = 0$ and inspecing their concavity.  Fixed points in the potential function which are concave-up are stable.

To observe the shape of the extended HKB potential function, consider a set of points and some fixed parameter values.

```python
A01 = 1
A10 = 1
B01 = 1
B10 = 1
d_omega = 0

x = np.arange(-5, 5, 0.1)
V = -d_omega*x - (A01 + A10)*np.cos(x) - (B01 + B10)*np.cos(2*x)

fig, ax = plt.subplots(figsize = (12, 6))

ax.set(xlabel = 'phi', ylabel = 'V')
ax.plot(x, V)
```

For $\delta\omega = 0$, we can readily identify the stable fixed points of the HKB model.  There is the minimum at the center about $\phi = 0$ corresponding to the in-phase state, and there are two minima at $\phi = -\pi$ and $\phi = \pi$ corresponding to the antiphase state.

As $\frac{b}{a}$ decreases, the two minima for the anti-phase fixed point rise, until after crossing the critical value $\frac{b}{a} = \frac{1}{4}$ the two minima for the anti-phase fixed point disappear completely.  Hence, stability for this point is lost and only in-phase motion becomes stable.

Observe the anti-phase minima rise.

```python
a = 2
b = 1
d_omega = 0

x = np.arange(-5, 5, 0.1)
V = -d_omega*x - a*np.cos(x) - b*np.cos(2*x)

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(x, V)
```

And see them disappear completely.

```python
a = 4
b = 1
d_omega = 0

x = np.arange(-5, 5, 0.1)
V = -d_omega*x - a*np.cos(x) - b*np.cos(2*x)

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(x, V)
```

Now, adding the frequency difference $\delta\omega$ corresponds to broken symmetry.  Hence, the potential function becomes asymmetrical, with the degree of the asymmetry corresponding to the size of the frequency difference $\delta\omega$.

Try with $\delta\omega = 1$

```python
a = 1
b = 1
d_omega = 1

x = np.arange(-5, 5, 0.1)
V = -d_omega*x - a*np.cos(x) - b*np.cos(2*x)

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(x, V)
```

We see that cranking up $\delta\omega$ results in the potential curve becoming more and more like a straight line and less like a linear combination of cosines, resulting in a loss of the local minima and hence a loss of the stable in-phase and antiphase fixed points.  

Finally, note that even for large frequency differences, if the coupling parameters are large enough then the potential function will still look like a combination of cosines, as the influence of the cosine terms grows as we increase their coefficients.  This explains the preservation of the stable fixed points despite a large frequency difference that occurs when the coupling constants are large enough.

## Measuring Synchronization

Synchronization of the oscillators can be measured by computing the cluster phase or Kuramoto order parameter.  The cluster phase at time $t$ in its complex form $q'(t) \in \mathbb{C}$ is defined as 

\begin{equation}
    q'(t) := \frac{1}{N} \sum \limits _{k = 1} ^{N} e^{i\theta_k(t)}
    \end{equation}

where $N$ is the number of oscillators in the system and $\theta_k(t)$ is the absolute phase of the kth oscillator at time $t$.

Given the complex form of the cluster phase, one can then compute its real form

\begin{equation}
    q(t) := {\rm atan2}(\operatorname{Im}[q'(t)], \operatorname{Re}[q'(t)])
    \end{equation}
    
where $q(t) \in [-\pi, \pi]$ measures the average phase of the entire group of oscillators at time $t$.

After computing the real form of the cluster phase, one can then take the relative phase between the cluster phase and the absolute phase of each oscillator at time $t$

\begin{equation}
    \phi_k(t) := \theta_k(t) - q(t)
    \end{equation}

This gives a measure of the difference between the oscillations of the kth agent the overall oscillations of the group for each time $t$.

One can then take the average value of this kth relative phase over the duration of the experiment $T$.  In its complex form $\bar \phi_k ' \in \mathbb{C}$, this average relative phase is defined as 

\begin{equation}
    \bar \phi_k ' := \frac {1}{T} \int_{0}^T e^{i\phi_k(t)}dt \approx \frac{1}{N_T} \sum \limits _{j = 1} ^{N_T} e^{i \phi_k(t_j)}
    \end{equation}

where $N_T$ is the total number of time points for which the discrete time series is measured.

The real form of the average relative phase between the cluster phase and each oscillator's phase, $\bar \phi_k \in [-\pi, \pi]$, is then given by

\begin{equation}
    \bar \phi_k := {\rm atan2}(\operatorname{Im}[\bar \phi_k '], \operatorname{Re}[\bar \phi_k '])
    \end{equation}

Once the relative phase between the group and the kth oscillator has been computed, one can compute the average degree of synchronization over the duration of the experiment between the kth oscillator and the group by taking the magnitude of the complex form of the average relative phase.

The degree of synchronization between the kth oscillator and the group is defined as 

\begin{equation}
    \rho_k := |\bar\phi_k'| \in [0, 1]
    \end{equation}
    
Values of $\rho_k$ closer to 1 denote higher synchronization, whereas values of $\rho_k$ closer to 0 denote lower synchronization.

It is also possible to measure the overall synchronization of the entire group at time $t$ with the degree of group synchronization at time $t$

\begin{equation} 
    \rho_g(t) := \frac{1}{N} \left |\sum \limits _{k = 1} ^{N} e^{i [\phi_k(t) - \bar \phi_k]} \right | \in [0, 1]
\end{equation}

The degree of group synchronization is defined similarly to the degree of synchronization between the kth oscillator and the group: values of $\rho_g (t)$ closer to 1 give higher levels of overall synchronization, while values of $\rho_g(t)$ closer to 0 give lower levels of group synchronization.

One can then average this quantity over the duration of the experiment to give the average degree of group synchronization

\begin{equation}
    \rho_g := \frac{1}{T} \int_{0}^T \rho_g(t) dt \approx \frac{1}{N_T} \sum \limits _{j = 1} ^{N_T} \rho_g(t_j)
    \end{equation}

Finally, we can measure the dyadic synchronization of two oscillators.  

If the absolute phases of oscillators k and k' are given by $\theta_k$ and $\theta_{k'}$ respectively, then the relative phase between oscillators k and k' at time $t$ is 

\begin{equation}
    \phi_{d_{kk'}} := \theta_k - \theta_{k'}
    \end{equation}
    
Then, the average dyadic degree of synchronization, or phase-locking value, over the duration of the experiment between oscillators k and k' is defined as

\begin{equation}
    \rho_{d_{kk'}} := \left | \frac{1}{T} \int_{0}^T e^{i \phi_{d_{kk'}}(t)} dt \right | \approx \left | \frac{1}{N_T} \sum \limits _{j = 1} ^{N_T} e^{i \phi_{d_{kk'}}(t_j)} \right | \in [0, 1]
    \end{equation}
    
As usual, values of $\rho_{d_{kk'}}$ closer to 1 give higher levels of dyadic synchronization, whereas values of $\rho_{d_{kk'}}$ closer to 0 give lower levels of dyadic synchronization.

Thus, these synchronization metrics allow us to measure the level of synchronization between each oscillator and the rest of the group, the overall level of synchronization of the entire group, and the level of synchronization between all oscillator dyads.  The experiment we wish to perform is to determine the effect that the number of oscillators $N$ has on the ability of the oscillators to synchronization.  So, our task is to compute values of these synchronization metrics for different values of $N$, and see how the synchronization metrics change as we vary $N$.

```python
def synch(theta):
    
    # Define the real and imaginary parts of the cluster phase.
    Im_q = np.sum(np.sin(theta), axis = 1)
    Re_q = np.sum(np.cos(theta), axis = 1)

    # Initialize the N_T-dimensional vector q.
    q = np.empty_like(t)

    # Set the ith entry of q to be equal to the atan2 of the real and imaginary parts of the cluster phase at time t_i.
    for i in range(len(t)):
        q[i] = np.arctan2(Im_q[i], Re_q[i])
        
    # Initialize the (N_T, N)-dimensional matrix of the relative phase.
    rel_phi_q = np.empty_like(theta)

    # Set the ith column of the relative phase matrix to be equal to the difference between the ith oscillator and q for each 
    # time point.
    for i in range(N):
        rel_phi_q[:, i] = theta[:, i] - q
        
    # Define the real and imaginary parts of the time-averaged relative phase between each oscillator and the cluster phase.
    Im_phi_bar = np.sum(np.sin(rel_phi_q), axis = 0)
    Re_phi_bar = np.sum(np.cos(rel_phi_q), axis = 0)

    # Initialize the real form of the average relative phase between the cluster phase and each oscillator.
    phi_bar = np.empty((N, 1))

    # Define the real form of the average relative phase between the cluster phase and each oscillator.
    for i in range(0, N):
        phi_bar[i] = np.arctan2(Im_phi_bar[i], Re_phi_bar[i])

    # Initialize the N-dimensional vector rho giving the degree of synchronization for each oscillator with respect to the 
    # group.
    rho = np.empty((N, 1))

    # Set each of the entries of rho.
    for i in range(0, N):
        rho[i] = (1/N_T)*np.sqrt((Im_phi_bar[i])**2 + (Re_phi_bar[i])**2)
    
    print('The degree of synchronization between each oscillator and the group is \n', rho)
    
    
    # Define the real and imaginary parts of this quantity.
    Im_rel_phi_phi_bar = np.sum(np.sin(rel_phi_q - np.transpose(phi_bar)), axis = 1)
    Re_rel_phi_phi_bar = np.sum(np.cos(rel_phi_q - np.transpose(phi_bar)), axis = 1)

    # Initialize the N_T-dimensional vector for rho_g(t).
    rho_g = np.empty_like(t)

    # Set each entry of rho_g(t) to be equal to its value at each time point.
    for i in range(len(t)):
        rho_g[i] = (1/N)*np.sqrt((Im_rel_phi_phi_bar[i])**2 + (Re_rel_phi_phi_bar[i])**2)
    
    # Now take the average value of rho_g(t) over the duration of the experiment.
    avg_rho_g = (1/N_T)*np.sum(rho_g, axis = 0)
    
    print('The average degree of group synchronization is \n', avg_rho_g)
    
    
    # Initialize the array for the dyadic relative phases at each time point.
    phi_d = np.empty([N_T, N, N])

    # Set the entries to be the relative phase for each time point.
    for i in range (0, N_T):
        phi_d[i] = np.subtract.outer(theta[i], theta[i])
    
    # Define the real and imaginary parts of phase locking value.
    Im_phi_d = np.sum(np.sin(phi_d), axis = 0)
    Re_phi_d = np.sum(np.cos(phi_d), axis = 0)  

    # Initialize an array for the dyadic synchronization.
    rho_d = np.empty([N, N])

    # Set each entry in rho_d to be the average dyadic synchronization over the duration of the experiment.
    for i in range(0, N):
        for j in range(0, N):
            rho_d[i, j] = (1/N_T)*np.sqrt((Im_phi_d[i, j])**2 + (Re_phi_d[i, j])**2)
            
    print('The dyadic degree of synchronization matrix is \n', rho_d)
```
